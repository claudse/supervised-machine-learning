# Predicting the House Prices

## Problem

In this project we want to predict the house prices using advanced regression techniques. The dataset we use contains 
79 variables, which is a large number of features. Therefore feature selection will be important for our model, since we want to prevent overfitting.

## Dataset

The data is from "The Ames Housing dataset" which was compiled by Dean De Cock and  can be found on Kaggle: 
    <a href="https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data">
    www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data</a>

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_data = pd.read_csv('train.csv')
train_data.describe(include='all') # 'all' includes the categorical data 
```

We define the features by X and the output by y, which contains the Sale Price of the house.

```python
X = train_data.drop(['Id', 'SalePrice'], axis=1)
y = train_data['SalePrice']
```

We check if all NaN values are removed:

### 2. Encoding Categorical Data to One Hot Encoding

Our next step is to encode the categorical data into integer values, since String values will cause problems in a regression method. We use one hot encoding for variables which don't have a ranked order, or for variables where we the order might not be very important.

```python
ohe_df_MSZoning = pd.get_dummies(X['MSZoning'])
ohe_df_Street = pd.get_dummies(X['Street'])
ohe_df_Alley = pd.get_dummies(X['Alley'])
ohe_df_LandContour = pd.get_dummies(X['LandContour'])
ohe_df_Utilities = pd.get_dummies(X['Utilities'])
ohe_df_LotConfig = pd.get_dummies(X['LotConfig'])
ohe_df_Neighborhood = pd.get_dummies(X['Neighborhood'])
ohe_df_Condition1 = pd.get_dummies(X['Condition1'])
ohe_df_Condition2 = pd.get_dummies(X['Condition2'])
ohe_df_BldgType = pd.get_dummies(X['BldgType'])
ohe_df_RoofStyle = pd.get_dummies(X['RoofStyle'])
ohe_df_RoofMatl = pd.get_dummies(X['RoofMatl'])
ohe_df_Exterior1st = pd.get_dummies(X['Exterior1st'])
ohe_df_Exterior2nd = pd.get_dummies(X['Exterior2nd'])
ohe_df_MasVnrType = pd.get_dummies(X['MasVnrType'])
ohe_df_Foundation = pd.get_dummies(X['Foundation'])
ohe_df_Heating = pd.get_dummies(X['Heating'])
ohe_df_CentralAir = pd.get_dummies(X['CentralAir'])
ohe_df_Electrical = pd.get_dummies(X['Electrical'])
ohe_df_GarageType = pd.get_dummies(X['GarageType'])
ohe_df_GarageFinish = pd.get_dummies(X['GarageFinish'])
ohe_df_GarageCond = pd.get_dummies(X['GarageCond'])
ohe_df_PavedDrive = pd.get_dummies(X['PavedDrive'])
ohe_df_PoolQC = pd.get_dummies(X['PoolQC'])
ohe_df_Fence = pd.get_dummies(X['Fence'])
ohe_df_MiscFeature = pd.get_dummies(X['MiscFeature'])
ohe_df_SaleType = pd.get_dummies(X['SaleType'])
ohe_df_SaleCondition = pd.get_dummies(X['SaleCondition'])

X = pd.concat([X, ohe_df_MSZoning, ohe_df_Street, ohe_df_Alley, ohe_df_LandContour, ohe_df_Utilities, 
               ohe_df_LotConfig, ohe_df_Neighborhood, ohe_df_Condition1, ohe_df_Condition2, 
               ohe_df_BldgType, ohe_df_RoofStyle, ohe_df_RoofMatl, ohe_df_Exterior1st, 
               ohe_df_Exterior2nd, ohe_df_MasVnrType, ohe_df_Foundation, ohe_df_Heating, 
               ohe_df_CentralAir, ohe_df_Electrical, ohe_df_GarageType, ohe_df_GarageFinish, 
               ohe_df_GarageCond, ohe_df_PavedDrive, ohe_df_PoolQC, ohe_df_Fence, ohe_df_MiscFeature, 
               ohe_df_SaleType, ohe_df_SaleCondition], 
              axis=1).drop(['MSZoning', 'Street', 'Alley', 'LandContour', 'Utilities', 'LotConfig', 
                            'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'RoofStyle',
                            'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation', 
                            'Heating', 'CentralAir', 'Electrical', 'GarageType', 'GarageFinish', 
                            'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 
                            'SaleCondition'], axis=1)
```

### 3. Encoding Ordinal Categorical Data with Label Encoder

The label encoder maps the categorical values to integer values starting with zero. It will encode alphabetically ordered. To keep the right order, we will replace some features with an according letter.

```python
X['LotShape'] = X['LotShape'].replace('Reg','IR0') # 'IR0' for alphabetical order

X['HouseStyle'] = X['HouseStyle'].replace('1Story','A') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('1.5Fin','B') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('1.5Unf','C') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('2Story','D') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('2.5Fin','E') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('2.5Unf','F') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('SFoyer','G') # for alphabetical order
X['HouseStyle']  = X['HouseStyle'].replace('SLvl','H') # for alphabetical order

X['ExterQual'] = X['ExterQual'].replace('TA','Ge') #'Ge' for alphabetical order

X['ExterCond'] = X['ExterCond'].replace('TA','Ge') #'Ge' for alphabetical order

X['BsmtQual'] = X['BsmtQual'].replace('TA','Ge') #'Ge' for alphabetical order
X['BsmtQual'] = X['BsmtQual'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last

X['BsmtCond'] = X['BsmtCond'].replace('TA','Ge') #'Ge' for alphabetical order
X['BsmtCond'] = X['BsmtCond'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last

X['BsmtExposure'] = X['BsmtExposure'].replace('Av','Ge') #'Ge' for alphabetical order
X['BsmtExposure'] = X['BsmtExposure'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last

X['BsmtFinType1'] = X['BsmtFinType1'].replace('GLQ','A') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('ALQ','B') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('BLQ','C') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('Rec','D') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('LwQ','E') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('Unf','F') # for alphabetical order
X['BsmtFinType1'] = X['BsmtFinType1'].replace('NA','G') # for alphabetical order

X['BsmtFinType2'] = X['BsmtFinType2'].replace('GLQ','A') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('ALQ','B') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('BLQ','C') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('Rec','D') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('LwQ','E') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('Unf','F') # for alphabetical order
X['BsmtFinType2'] = X['BsmtFinType2'].replace('NA','G') # or alphabetical order

X['HeatingQC'] = X['HeatingQC'].replace('TA','Ge') #'Ge' for alphabetical order

X['KitchenQual'] = X['KitchenQual'].replace('TA','Ge') #'Ge' for alphabetical order
X['KitchenQual'] = X['KitchenQual'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last

X['Functional'] = X['Functional'].replace('Typ','A') # for alphabetical order
X['Functional'] = X['Functional'].replace('Min1','B1') # for alphabetical order
X['Functional'] = X['Functional'].replace('Min2','B2') # for alphabetical order
X['Functional'] = X['Functional'].replace('Mod','C') # for alphabetical order
X['Functional'] = X['Functional'].replace('Maj1','D1') # for alphabetical order
X['Functional'] = X['Functional'].replace('Maj2','D2') # for alphabetical order
X['Functional'] = X['Functional'].replace('Sev','E') # for alphabetical order
X['Functional'] = X['Functional'].replace('Sal','F') # for alphabetical order

X['FireplaceQu']=X['FireplaceQu'].replace('TA','Ge') #'Ge' for alphabetical order
X['FireplaceQu']=X['FireplaceQu'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last


X['GarageQual'] = X['GarageQual'].replace('TA','Ge') #'Ge' for alphabetical order
X['GarageQual'] = X['GarageQual'].replace('NA','Pp') #'Pp' to make sure it is alphabetically the last
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

X['KitchenQual'] = le.fit_transform(X['KitchenQual'])
X['LandSlope'] = le.fit_transform(X['LandSlope'])
X['LotShape'] = le.fit_transform(X['LotShape'])
X['HeatingQC'] = le.fit_transform(X['HeatingQC'])
X['ExterQual'] = le.fit_transform(X['ExterQual'])
X['ExterCond'] = le.fit_transform(X['ExterCond'])
X['GarageQual'] = le.fit_transform(X['GarageQual'])
X['FireplaceQu'] = le.fit_transform(X['FireplaceQu'])
X['BsmtFinType1'] = le.fit_transform(X['BsmtFinType1'])
X['BsmtFinType2'] = le.fit_transform(X['BsmtFinType2'])
X['BsmtExposure'] = le.fit_transform(X['BsmtExposure'])
X['BsmtQual'] = le.fit_transform(X['BsmtQual'])
X['BsmtCond'] = le.fit_transform(X['BsmtCond'])
X['HouseStyle'] = le.fit_transform(X['HouseStyle'])
X['Functional'] = le.fit_transform(X['Functional'])
```

## Feature Selection

We want to avoid overfitting in our model by using to many features. With the analysis above we have investigated which features are more informative for the prediction of the house price. We will use the features have a correlation with the Sale Price. Also, we want to use the features with strong negative correlations. But we will leave out 'GarageArea' and 'GarageYrBlt' because they are strongly correlated to 'YearBuilt' and 'GarageCars'. The computation of the model will have a better performance, if we dont use to much memory.

```python
X_selected_f = X[[
'OverallQual',
'GrLivArea',
'GarageCars',
'TotalBsmtSF',
'1stFlrSF',
'FullBath',
'TotRmsAbvGrd',
'YearBuilt',
'YearRemodAdd',
'MasVnrArea',
'Fireplaces',
'ExterCond',
'ExterQual',
'BsmtQual',
'KitchenQual',
'FireplaceQu',
'HeatingQC',
'PConc',
'Fin',
'NridgHt'
]].copy()

```

## Model Training

### Decision Tree Regressor

Now we will make a regression using the Decision Tree Regressor from sklearn. We will divide the data in a training set and a test set. Finally we will compute the mean squared error of the regression for the test data. To visualize we make a plot of the predicted Sale Price and the true Sale Price.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected_f, y, test_size=0.23, random_state=42)
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
plt.figure(figsize=(5,5))
plt.scatter(y_pred, y_test, s=0.8)
plt.plot([0,700000],[0,700000], color='r')
plt.xlabel('y_pred')
plt.ylabel('y_test')
meanSquaredError=mean_squared_error(y_test, y_pred)
print("MSE:", meanSquaredError)

rootMeanSquaredError = np.sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
```

### Recursive Feature Elimination

RFE is method to reduce the numbers of features. It looks for the most informative features for a provided supervised machine learning method. The desired number of features has to be given. Let's see if it will select similar features.

```python
from sklearn.feature_selection import RFE
```

```python
X_train_rfe, X_test_rfe, y_train_rfe, y_test_rfe = train_test_split(X, y, test_size=0.23, random_state=42)

rfe = RFE(estimator=DecisionTreeRegressor(), n_features_to_select=20)
rfe.fit(X_train_rfe, y_train_rfe)
print(X.columns.values[rfe.support_])
```

```python
y_pred_rfe = rfe.predict(X_test_rfe)

plt.figure(figsize=(5,5))
plt.scatter(y_pred_rfe, y_test_rfe, s=0.8)
plt.plot([0,700000],[0,700000], color='r')
plt.xlabel('y_pred_rfe')
plt.ylabel('y_test_rfe')
meanSquaredError_rfe = mean_squared_error(y_test_rfe, y_pred_rfe)
print("MSE with RFE:", meanSquaredError_rfe)

rootMeanSquaredError_rfe = np.sqrt(meanSquaredError_rfe)
print("RMSE with RFE:", rootMeanSquaredError_rfe)
```

### Gradient Boosting Regressor

Let's see if we can get improve our prediction results using another regression method. We want to see how the Gradient Boosting Regressor performs. We will see that the MSE will even decrease.

```python
from sklearn import ensemble
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}
```

```python
reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(X_train, y_train)
```

```python
y_pred_gradient = reg.predict(X_test)
plt.figure(figsize=(5,5))
plt.scatter(y_pred_gradient, y_test, s=0.8)
plt.plot([0,700000],[0,700000], color='r')
plt.xlabel('y_pred_gradient')
plt.ylabel('y_test')
meanSquaredError_gradient = mean_squared_error(y_test, y_pred_gradient)
print("MSE with Gradient Boosting Regressor:", meanSquaredError_gradient)

rootMeanSquaredError_gradient = np.sqrt(meanSquaredError_gradient)
print("RMSE with Gradient Boosting Regressor:", rootMeanSquaredError_gradient)
```
